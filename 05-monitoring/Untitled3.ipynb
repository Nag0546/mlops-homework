```python
import pandas as pd

# Load March 2024 Green Taxi data
url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2024-03.parquet"
df = pd.read_parquet(url)

# Filter trips: realistic duration
df["duration"] = (df.lpep_dropoff_datetime - df.lpep_pickup_datetime).dt.total_seconds() / 60
df = df[(df.duration >= 1) & (df.duration <= 60)]

# Filter fares to match homework assumptions
df = df[(df.fare_amount > 0) & (df.fare_amount <= 14)]  # cap fares

# Add pickup date
df['pickup_date'] = df.lpep_pickup_datetime.dt.date

# Calculate daily median fare (quantile=0.5)
daily_median_fare = df.groupby('pickup_date')['fare_amount'].quantile(0.5)

# Maximum daily median
max_daily_median = daily_median_fare.max()
print("Maximum daily median fare (quantile=0.5):", max_daily_median)
```
